```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
```

```python
DATA_PATH = '../data/Environment_Temperature_change_E_All_Data_NOFLAG.csv'
```

```python
data_frame = pd.read_csv(DATA_PATH, encoding='cp1252')
print(data_frame.shape)
```

(codul poate pare asemanator cu cel din data_analysis, dar o sa arate alte statistici, trust me :D)

```python
year_columns = ['Y' + str(i) for i in range(1961,2020)]
```

<h3> Indexarea lunilor este foarte mare, si foloseste mai multi indecsi decat este nevoie.

```python
int(max(data_frame["Months Code"].values)) - int(min(data_frame["Months Code"].values)) + 1
```

```python
le_months = LabelEncoder()
le_months.fit(data_frame["Months"])
data_frame["Months"] = le_months.transform(data_frame["Months"].values)

data_frame.iloc[:10, : 10]
```

```python
max(data_frame["Months"].values)
```

<h3> Se observa ca sunt mai putini indecsi folositi, si putem scapa de randul de Months Code. Label Encoder-ul o sa tina minte label-urile fiecarui index.

```python
data_frame = data_frame.drop(columns=["Months Code"])
```

<h3> Putem aplica aceeasi chestie si pentru tari

```python
le_countries = LabelEncoder()
le_countries.fit(data_frame["Area"])
data_frame["Area"] = le_countries.transform(data_frame["Area"].values)

data_frame.iloc[:10, : 10]
```

<h3> Aplicam aceeasi logica pentru eliminarea coloanei Area Code.

```python
data_frame = data_frame.drop(columns=["Area Code"])
```

<h3> Aplicam Standard Scaling pe coloanele anilor

```python
data_frame[year_columns].iloc[:10, :10]
```

```python
data_frame[year_columns] = StandardScaler().fit_transform(data_frame[year_columns].values)
data_frame[year_columns].iloc[:10, :10]
```

<h3> Pentru ca avem foarte multi ani, o sa incercam sa reducem unii dintre ei, cu o metoda clasica, fiind PCA

```python
# aceasta parte trebuie discutata si cu cealalta echipa
for column in year_columns:
    data_frame[column] = data_frame[column].replace(np.nan, data_frame[column].mean())

pca = PCA(n_components='mle')

new_features = pca.fit_transform(data_frame[year_columns])

print("Original number of features: ", data_frame.shape[1])
print("Reduced number of features: ", new_features.shape[1])
```

<h3> Se poate observa ca se reduc 9 feature-uri din anii fiecarei dati.
